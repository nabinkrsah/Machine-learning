```python
# to calculate the pollutant index for so2(sulphur dioxide)

def cal_SOi(so2):
    si=0
    if (so2<=40):
     si= so2*(50/40)
    elif (so2>40 and so2<=80):
     si= 50+(so2-40)*(50/40)
    elif (so2>80 and so2<=380):
     si= 100+(so2-80)*(100/300)
    elif (so2>380 and so2<=800):
     si= 200+(so2-380)*(100/420)
    elif (so2>800 and so2<=1600):
     si= 300+(so2-800)*(100/800)
    elif (so2>1600):
     si= 400+(so2-1600)*(100/800)
    return si
df['SOi']=df['so2'].apply(cal_SOi)
data= df[['so2','SOi']]
data.head()
```

```python
# to calculate the pollutant index for no2(nitrogen dioxide)
def cal_Noi(no2):
    ni=0
    if(no2<=40):
     ni= no2*50/40
    elif(no2>40 and no2<=80):
     ni= 50+(no2-40)*(50/40)
    elif(no2>80 and no2<=180):
     ni= 100+(no2-80)*(100/100)
    elif(no2>180 and no2<=280):
     ni= 200+(no2-180)*(100/100)
    elif(no2>280 and no2<=400):
     ni= 300+(no2-280)*(100/120)
    else:
     ni= 400+(no2-400)*(100/120)
    return ni
df['Noi']=df['no2'].apply(cal_Noi)
data= df[['no2','Noi']]
data.head()
```

```python
# to calculate the pollutant index for RSPMI
def cal_RSPMI(rspm):
    rpi=0
    if(rpi<=30):
     rpi=rpi*50/30
    elif(rpi>30 and rpi<=60):
     rpi=50+(rpi-30)*50/30
    elif(rpi>60 and rpi<=90):
     rpi=100+(rpi-60)*100/30
    elif(rpi>90 and rpi<=120):
     rpi=200+(rpi-90)*100/30
    elif(rpi>120 and rpi<=250):
     rpi=300+(rpi-120)*(100/130)
    else:
     rpi=400+(rpi-250)*(100/130)
    return rpi
df['Rpi']=df['rspm'].apply(cal_RSPMI)
data= df[['rspm','Rpi']]
data.head()
```

```python
# to calculate the pollutant index for spm(suspended particulate matter)
def cal_SPMi(spm):
    spi=0
    if(spm<=50):
     spi=spm*50/50
    elif(spm>50 and spm<=100):
     spi=50+(spm-50)*(50/50)
    elif(spm>100 and spm<=250):
     spi= 100+(spm-100)*(100/150)
    elif(spm>250 and spm<=350):
     spi=200+(spm-250)*(100/100)
    elif(spm>350 and spm<=430):
     spi=300+(spm-350)*(100/80)
    else:
     spi=400+(spm-430)*(100/430)
    return spi
   
df['SPMi']=df['spm'].apply(cal_SPMi)
data= df[['spm','SPMi']]
data.head()
```

```python
#to calculate the air quality index.
def cal_aqi(si,ni,rspmi,spmi):
    aqi=0
    if(si>ni and si>rspmi and si>spmi):
     aqi=si
    if(ni>si and ni>rspmi and ni>spmi):
     aqi=ni
    if(rspmi>si and rspmi>ni and rspmi>spmi):
     aqi=rspmi
    if(spmi>si and spmi>ni and spmi>rspmi):
     aqi=spmi
    return aqi

df['AQI']=df.apply(lambda x:cal_aqi(x['SOi'],x['Noi'],x['Rpi'],x['SPMi']),axis=1)
data= df[['state','SOi','Noi','Rpi','SPMi','AQI']]
data.head()
```

```python
# Using threshold values to classify a particular values as good, moderate, 
# poor, unhealthy, very unhealthy and Hazardous
def AQI_Range(x):
    if x<=50:
        return "Good"
    elif x>50 and x<=100:
        return "Moderate"
    elif x>100 and x<=200:
        return "Poor"
    elif x>200 and x<=300:
        return "Unhealthy"
    elif x>300 and x<=400:
        return "Very unhealthy"
    elif x>400:
        return "Hazardous"

df['AQI_Range'] = df['AQI'] .apply(AQI_Range)
df.head()
```

```python
# Extract the target variable (AQI) for outlier detection
target_variable = 'AQI'
data_for_outliers = df[target_variable]

# Calculate the IQR (Interquartile Range)
Q1 = data_for_outliers.quantile(0.25)
Q3 = data_for_outliers.quantile(0.75)
IQR = Q3 - Q1

# Define a threshold for identifying outliers
outlier_threshold = 1.5

# Identify outliers using the IQR method
outliers = (data_for_outliers < Q1 - outlier_threshold * IQR) | (data_for_outliers > Q3 + outlier_threshold * IQR)

# Print the indices of the identified outliers
print("Indices of Outliers:")
print(df[outliers].index)

```

```python
from scipy import stats
# Extract numerical columns for outlier removal
numerical_columns = ['SOi', 'Noi', 'Rpi', 'SPMi', 'AQI']

# Calculate Z-scores for each numerical column
z_scores = stats.zscore(df[numerical_columns])

# Set a threshold for Z-scores (e.g., 3 standard deviations)
threshold = 3

# Identify outliers based on the threshold
outliers = (abs(z_scores) > threshold).any(axis=1)

# Remove outliers from the dataframe
df_no_outliers = df[~outliers]

# Display the dataframe without outliers
print("DataFrame without Outliers:")
df_no_outliers
```

```python
from sklearn.preprocessing import MinMaxScaler
# Extract numerical columns for normalization
numerical_columns = ['SOi', 'Noi', 'Rpi', 'SPMi', 'AQI']

# Create a MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numerical columns
df_no_outliers[numerical_columns] = scaler.fit_transform(df_no_outliers[numerical_columns])

# Display the normalized dataframe
print("Normalized DataFrame:")
df_no_outliers
```
