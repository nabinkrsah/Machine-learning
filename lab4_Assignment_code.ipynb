```python
# to calculate the pollutant index for so2(sulphur dioxide)

def cal_SOi(so2):
    si=0
    if (so2<=40):
     si= so2*(50/40)
    elif (so2>40 and so2<=80):
     si= 50+(so2-40)*(50/40)
    elif (so2>80 and so2<=380):
     si= 100+(so2-80)*(100/300)
    elif (so2>380 and so2<=800):
     si= 200+(so2-380)*(100/420)
    elif (so2>800 and so2<=1600):
     si= 300+(so2-800)*(100/800)
    elif (so2>1600):
     si= 400+(so2-1600)*(100/800)
    return si
df['SOi']=df['so2'].apply(cal_SOi)
data= df[['so2','SOi']]
data.head()
```

```python
# to calculate the pollutant index for no2(nitrogen dioxide)
def cal_Noi(no2):
    ni=0
    if(no2<=40):
     ni= no2*50/40
    elif(no2>40 and no2<=80):
     ni= 50+(no2-40)*(50/40)
    elif(no2>80 and no2<=180):
     ni= 100+(no2-80)*(100/100)
    elif(no2>180 and no2<=280):
     ni= 200+(no2-180)*(100/100)
    elif(no2>280 and no2<=400):
     ni= 300+(no2-280)*(100/120)
    else:
     ni= 400+(no2-400)*(100/120)
    return ni
df['Noi']=df['no2'].apply(cal_Noi)
data= df[['no2','Noi']]
data.head()
```

```python
# to calculate the pollutant index for RSPMI
def cal_RSPMI(rspm):
    rpi=0
    if(rpi<=30):
     rpi=rpi*50/30
    elif(rpi>30 and rpi<=60):
     rpi=50+(rpi-30)*50/30
    elif(rpi>60 and rpi<=90):
     rpi=100+(rpi-60)*100/30
    elif(rpi>90 and rpi<=120):
     rpi=200+(rpi-90)*100/30
    elif(rpi>120 and rpi<=250):
     rpi=300+(rpi-120)*(100/130)
    else:
     rpi=400+(rpi-250)*(100/130)
    return rpi
df['Rpi']=df['rspm'].apply(cal_RSPMI)
data= df[['rspm','Rpi']]
data.head()
```

```python
# to calculate the pollutant index for spm(suspended particulate matter)
def cal_SPMi(spm):
    spi=0
    if(spm<=50):
     spi=spm*50/50
    elif(spm>50 and spm<=100):
     spi=50+(spm-50)*(50/50)
    elif(spm>100 and spm<=250):
     spi= 100+(spm-100)*(100/150)
    elif(spm>250 and spm<=350):
     spi=200+(spm-250)*(100/100)
    elif(spm>350 and spm<=430):
     spi=300+(spm-350)*(100/80)
    else:
     spi=400+(spm-430)*(100/430)
    return spi
   
df['SPMi']=df['spm'].apply(cal_SPMi)
data= df[['spm','SPMi']]
data.head()
```

```python
# Caluclating the Air Quality Index.
#to calculate the air quality index.
def cal_aqi(si,ni,rspmi,spmi):
    aqi=0
    if(si>ni and si>rspmi and si>spmi):
     aqi=si
    if(ni>si and ni>rspmi and ni>spmi):
     aqi=ni
    if(rspmi>si and rspmi>ni and rspmi>spmi):
     aqi=rspmi
    if(spmi>si and spmi>ni and spmi>rspmi):
     aqi=spmi
    return aqi

df['AQI']=df.apply(lambda x:cal_aqi(x['SOi'],x['Noi'],x['Rpi'],x['SPMi']),axis=1)
data= df[['state','SOi','Noi','Rpi','SPMi','AQI']]
data.head()
```

```python
# Using threshold values to classify a particular values as good, moderate, 
# poor, unhealthy, very unhealthy and Hazardous
def AQI_Range(x):
    if x<=50:
        return "Good"
    elif x>50 and x<=100:
        return "Moderate"
    elif x>100 and x<=200:
        return "Poor"
    elif x>200 and x<=300:
        return "Unhealthy"
    elif x>300 and x<=400:
        return "Very unhealthy"
    elif x>400:
        return "Hazardous"

df['AQI_Range'] = df['AQI'] .apply(AQI_Range)
df.head()
```

```python
#Evaluate the intraclass spread and interclass distances between the classes in your dataset. If 
#our data deals with multiple classes, you can take any two classes. Steps below (refer below 
#diagram for understanding):
#Calculate the mean for each class (also called as class centroid)
#Calculate spread (standard deviation) for each class
#Calculate the distance between mean vectors between classes
```

```python
# Group the data by the "state" column to create classes
class_groups = df.groupby('state')

# Initialize lists to store class centroids and spreads
class_centroids = []
class_spreads = []

# Calculate mean and spread for each class
for name, group in class_groups:
    # Calculate mean (centroid) for each numeric attribute
    centroid = group[['so2', 'no2', 'rspm', 'spm', 'pm2_5']].mean()
    class_centroids.append((name, centroid))
    
    # Calculate standard deviation (spread) for each numeric attribute
    spread = group[['so2', 'no2', 'rspm', 'spm', 'pm2_5']].std()
    class_spreads.append((name, spread))
```

```python

# Calculate interclass distances
interclass_distances = []
for i in range(len(class_centroids)):
    for j in range(i+1, len(class_centroids)):
        class1_name, centroid1 = class_centroids[i]
        class2_name, centroid2 = class_centroids[j]
        distance = np.linalg.norm(centroid1 - centroid2)
        interclass_distances.append((class1_name, class2_name, distance))

# Print class centroids and spreads
for class_name, centroid in class_centroids:
    print(f"Class: {class_name}")
    print("Centroid (Mean):")
    print(centroid)
    print("\n")
```

```python

for class_name, spread in class_spreads:
    print(f"Class: {class_name}")
    print("Spread (Standard Deviation):")
    print(spread)
    print("\n")
```

```python

# Print interclass distances
for class1, class2, distance in interclass_distances:
    print(f"Interclass Distance between {class1} and {class2}: {distance}")
```

```python
import math

# Define a function to calculate entropy
def entropy(class_probs):
    entropy_value = 0
    for prob in class_probs:
        if prob != 0:
            entropy_value -= prob * math.log2(prob)
    return entropy_value

# Define a function to calculate Information Gain
def information_gain(data, attribute_name, class_name):
    # Calculate the entropy of the entire dataset
    total_entropy = entropy(data[class_name].value_counts(normalize=True))

    # Calculate the weighted average entropy after the split
    weighted_entropy = 0
    for attribute_value in data[attribute_name].unique():
        subset = data[data[attribute_name] == attribute_value]
        subset_probability = len(subset) / len(data)
        weighted_entropy += subset_probability * entropy(subset[class_name].value_counts(normalize=True))

    # Calculate Information Gain
    information_gain_value = total_entropy - weighted_entropy
    return information_gain_value



# Calculate Information Gain for each attribute
class_name = 'AQI_Range'
attributes = ['SOi','Noi','Rpi','SPMi']

information_gains = {}
for attribute in attributes:
    ig = information_gain(df, attribute, class_name)
    information_gains[attribute] = ig
    
    root_node1='SOi'
    root_node2='Noi'
    root_node3='Rpi'
    root_node4='SPMi'
    
print(f"The root node for the decision tree is '{root_node1}' with Information Gain of {information_gains[root_node1]}")
print(f"The root node for the decision tree is '{root_node2}' with Information Gain of {information_gains[root_node2]}")
print(f"The root node for the decision tree is '{root_node3}' with Information Gain of {information_gains[root_node3]}")
print(f"The root node for the decision tree is '{root_node4}' with Information Gain of {information_gains[root_node4]}")
print("now the with the highest information gain")
# Find the attribute with the highest Information Gain (root node)
root_node = max(information_gains, key=information_gains.get)

print(f"The root node for the decision tree is '{root_node}' with Information Gain of {information_gains[root_node]}")
```
